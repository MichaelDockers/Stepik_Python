## Task "Песочные часы"

Напишите программу с использованием рекурсивной функции, которая выводит изображение песочных часов, составленное из цифр 1, 2, 3, и 4:

1111111111111111    # 16 раз  
  222222222222      # 12 раз  
    33333333        # 8 раз  
      4444          # 4 раза  
    33333333        # 8 раз  
  222222222222      # 12 раз  
1111111111111111    # 16 раз  

**Формат входных данных**

На вход программе ничего не подается.

**Формат выходных данных**

Программа должна вывести изображение песочных часов, составленное из цифр 1, 2, 3, и 4, представленное в условии задачи.

**Примечание.** Количество цифр в каждой строке, указанное в комментариях, выводить не нужно.

```python
def sand_clock():
    def rec(n):
        
        if n > 1:
            print(' ' * 2 * (4 - n) + str(5 - n) * (4 * n))
            rec(n - 1)
        print(' ' * 2 * (4 - n) + str(5 - n) * (4 * n))
    rec(4)

sand_clock()
    
```
